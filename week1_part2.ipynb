## Data <a name= "data" ></a>

City of interest: **Paris** 

Based on definition of the problem, the following factors need to be considered:


**Data 1** Location of neighborhoods in Paris
Paris had 20 neighborhoods which is also called as arrondissement. The location data for the neighborhoods are readily available from the French Government website. https://www.data.gouv.fr/en/datasets/arrondissements-1/

**Data 2** Top Attractions
Distance of neighborhood from most popular tourists locations such as Notre Dame, Eiffel Tower, Louvre Museum,etc to attract more tourists to the restaurant. The data of the most popular destinations in Paris are obatined from US News. https://travel.usnews.com/Paris_France/Things_To_Do/

**Data 3** Demographics of Paris
* Population of paris
* Demographics of the population in the neighborhood 
This data will be found from the following websites
https://en.wikipedia.org/wiki/Paris
https://en.wikipedia.org/wiki/Demographics_of_Paris

**Data 4** Asian restaurants
 The location of existing asian restaurants and asian fusion restaurants(if any) in the neighborhood will be required. To gather this information ** Foursquare API** will be utilized.

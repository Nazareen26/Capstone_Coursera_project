# Capstone Project - The Battle of the Neighborhoods (Week 1)


### Asian Fusion Restaurant in Paris, France


## Introduction/Business Problem <a name="introduction "></a>


In this project, an optimal location for a restaurant in the city of Paris will be found. Paris is one of the most travelled cities in the world which includes several diverse and multicultural travelers from different countries. More than 1.8 million Chinese travellers, 1.1 million Japanese travelers and travellers from Asia visit Paris regularly. To cater to the needs of Asian travelers in high traffic locations like the Eiffel tower, this report will be targeted to find a location or locations for opening a Asian fusion restaurant in Paris,France. There are many restaurants in Paris and the possibilites of finding an area to establish a restaurant will be difficult. 

In this project, area for Asian fusion restaurant will be located based on high traffic areas closer to the Eiffel Tower, Notre Dame, Louvre museum , city center or the location that are not already crowded with other East Asian restaurants and areas where there are no Asian fusion restaurants in the vicinty. 

Techniques in data science will be used to generate the optimal location for the restaurant. Economic data and geographical location of the neighborhoods will be considered in the proccess of selecting a location for a new asian fusion restaurant in the city of Paris. 

Target Audience:
* Someone who wants to build new restaurant business in the city
* Tourists 
* Local residents
